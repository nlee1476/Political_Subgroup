### Neural Fitter Function
This function returns a dictionary of accuracy scores for each feature based on the training and test scores of the original version number. 

```python
def neural_fitter(train, test, list_of_feats):
    accuracy_dict = {}
    for x in list_of_feats:
        print(x)
        label_encoder_cols = [CategoricalEncoder(x)]
        data_encoder_cols = [
                         CategoricalEncoder('AGE8'),
                         CategoricalEncoder('PTYIDEO7'),
                         CategoricalEncoder('QRACEAI'),
                         CategoricalEncoder('SEN'),
                         CategoricalEncoder('EDUC18'),
                         CategoricalEncoder('HOU'),
                         CategoricalEncoder('GOV'),
                         CategoricalEncoder('SEX'),
                         CategoricalEncoder('PARTYID'),
                         CategoricalEncoder('PHIL3'),
                         CategoricalEncoder('SIZEPLAC'),
                         CategoricalEncoder('STANUM'),
                         CategoricalEncoder('CONTROLHOUSE'),
                         CategoricalEncoder('FAVDEM2'),
                         CategoricalEncoder('FAVREP2'),
        ]
    
        data_featurizer_cols = [EmbeddingFeaturizer('AGE8'),
                               EmbeddingFeaturizer ('PTYIDEO7'),
                               EmbeddingFeaturizer('QRACEAI'),
                               EmbeddingFeaturizer('SEN'),
                               EmbeddingFeaturizer('EDUC18'),
                               EmbeddingFeaturizer('HOU'),
                               EmbeddingFeaturizer('GOV'),
                               EmbeddingFeaturizer('SEX'),
                               EmbeddingFeaturizer('PARTYID'),
                               EmbeddingFeaturizer('PHIL3'),
                               EmbeddingFeaturizer('SIZEPLAC'),
                               EmbeddingFeaturizer('STANUM'),
                               EmbeddingFeaturizer('CONTROLHOUSE'),
                               EmbeddingFeaturizer('FAVDEM2'),
                               EmbeddingFeaturizer('FAVREP2') 
                               ]
        
        imputer = Imputer(
        data_featurizers= data_featurizer_cols,
        label_encoders= label_encoder_cols,
        data_encoders = data_encoder_cols,
        output_path = 'imputer_model')
    
        imputer.fit(train, num_epochs= 100)
        predictions, metrics = imputer.transform_and_compute_metrics(test)
        accuracy_dict[x] = metrics[x]['avg_accuracy']
        prob_dict = imputer.predict_proba(test)
        prob_dict_topk = imputer.predict_proba_top_k(test, top_k=5)
    return accuracy_dict
        
    
    
```

This function returns a dictionary of predictions and the probabiliy of each prediction within the testing data set

```python
def prob_dict_getter(train, test, feature):
    label_encoder_cols = [CategoricalEncoder(feature)]
    data_encoder_cols = [
                         CategoricalEncoder('AGE8'),
                         CategoricalEncoder('PTYIDEO7'),
                         CategoricalEncoder('QRACEAI'),
                         CategoricalEncoder('SEN'),
                         CategoricalEncoder('EDUC18'),
                         CategoricalEncoder('HOU'),
                         CategoricalEncoder('GOV'),
                         CategoricalEncoder('SEX'),
                         CategoricalEncoder('PARTYID'),
                         CategoricalEncoder('PHIL3'),
                         CategoricalEncoder('SIZEPLAC'),
                         CategoricalEncoder('STANUM'),
                
    ]
    
    data_featurizer_cols = [EmbeddingFeaturizer('AGE8'),
                               EmbeddingFeaturizer ('PTYIDEO7'),
                               EmbeddingFeaturizer('QRACEAI'),
                               EmbeddingFeaturizer('SEN'),
                               EmbeddingFeaturizer('EDUC18'),
                               EmbeddingFeaturizer('HOU'),
                               EmbeddingFeaturizer('GOV'),
                               EmbeddingFeaturizer('SEX'),
                               EmbeddingFeaturizer('PARTYID'),
                               EmbeddingFeaturizer('PHIL3'),
                               EmbeddingFeaturizer('SIZEPLAC'),
                               EmbeddingFeaturizer('STANUM'),
                           ]
        
    imputer = Imputer(
    data_featurizers= data_featurizer_cols,
    label_encoders= label_encoder_cols,
    data_encoders = data_encoder_cols,
    output_path = 'imputer_model')
    
    imputer.fit(train, num_epochs= 500)
    predictions, metrics = imputer.transform_and_compute_metrics(test)
    prob_dict = imputer.predict_proba(test)
    prob_dict_topk = imputer.predict_proba_top_k(test, top_k=5)
    return prob_dict_topk, predictions
```

### Function prediction_imputer
This function is used in the prediction maker to actually impute the predictions for the missing values into the dataframe

```python
def prediction_imputer(predictions, df, feature):
    preds = predictions[feature]
    df[feature] = preds
    return df
    
```

```python
def prediction_maker(train_df, predict_df_1, predict_df_2, predict_df_3, feature):
    label_encoder_cols = [CategoricalEncoder(feature)]
    data_encoder_cols = [
                         CategoricalEncoder('AGE8'),
                         CategoricalEncoder('PTYIDEO7'),
                         CategoricalEncoder('QRACEAI'),
                         CategoricalEncoder('SEN'),
                         CategoricalEncoder('EDUC18'),
                         CategoricalEncoder('HOU'),
                         CategoricalEncoder('GOV'),
                         CategoricalEncoder('SEX'),
                         CategoricalEncoder('PARTYID'),
                         CategoricalEncoder('PHIL3'),
                         CategoricalEncoder('SIZEPLAC'),
                         CategoricalEncoder('STANUM'),
                         CategoricalEncoder('FAVPARTY'),
                         CategoricalEncoder('PARTY'),
                         CategoricalEncoder('PARTYBYRACE'),
                         CategoricalEncoder('PTYIDEO'), 
                         CategoricalEncoder('AGEBYRACE08')
        
        
    ]
    
    data_featurizer_cols = [EmbeddingFeaturizer('AGE8'),
                               EmbeddingFeaturizer ('PTYIDEO7'),
                               EmbeddingFeaturizer('QRACEAI'),
                               EmbeddingFeaturizer('SEN'),
                               EmbeddingFeaturizer('EDUC18'),
                               EmbeddingFeaturizer('HOU'),
                               EmbeddingFeaturizer('GOV'),
                               EmbeddingFeaturizer('SEX'),
                               EmbeddingFeaturizer('PARTYID'),
                               EmbeddingFeaturizer('PHIL3'),
                               EmbeddingFeaturizer('SIZEPLAC'),
                               EmbeddingFeaturizer('STANUM'),
                               EmbeddingFeaturizer('FAVPARTY'),
                               EmbeddingFeaturizer('PARTY'), 
                               EmbeddingFeaturizer('PARTYBYRACE'), 
                               EmbeddingFeaturizer('PTYIDEO'),
                               EmbeddingFeaturizer('AGEBYRACE08')
        
        
                            
                           ]
        
    imputer = Imputer(
    data_featurizers= data_featurizer_cols,
    label_encoders= label_encoder_cols,
    data_encoders = data_encoder_cols,
    output_path = 'imputer_model')
    
    imputer.fit(train_df, num_epochs= 500)
    predictions1, metrics1 = imputer.transform_and_compute_metrics(predict_df_1)
    predictions2, metrics2 = imputer.transform_and_compute_metrics(predict_df_2)
    predictions3, metrics3 = imputer.transform_and_compute_metrics(predict_df_3)
    
    predict_df_1 = prediction_imputer(predictions1, predict_df_1, feature)
    predict_df_2 = prediction_imputer(predictions2, predict_df_2, feature)
    predict_df_3 = prediction_imputer(predictions3, predict_df_3, feature)
    
    return predict_df_1, predict_df_2, predict_df_3
```
